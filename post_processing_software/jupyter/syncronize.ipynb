```python

from cProfile import label
import csv
import matplotlib.pyplot as plt

def openR4ACSV(path):
    with open(path, 'r') as f:
        reader = csv.reader(f)
        header = next(reader)  # ヘッダー行を読み飛ばし

        data = []
        for row in reader:
            data.append(row)

    legends = [i for i in header]
    data = [[float(i) for i in row] for row in data]  # 文字列を数値に変換

    # X軸に時間を、Y軸にデータをセット
    d = [row[legends.index('Measured Frame')] for row in data]
    ratio = [row[legends.index('Ratio')] for row in data]
    curve = [row[legends.index('Fitted curve')] for row in data]
    # input = [row[legends.index('input')] for row in data]

    return d,ratio,curve
```

```python
def plotR4a(d,ratio,curve,xlim=None,ylim=None):
    
    plt.xlabel("Measured frames []")
    plt.ylabel("Relative average absolute angular acceleration []")  
    if xlim is not None: 
        plt.xlim(xlim)
    if ylim is not None: 
        plt.ylim(ylim)
    plt.plot(d,ratio)
    plt.plot(d,curve)
    # plt.plot(d,input)
    plt.show()
```

```python
for i in range(0,3):
    d,ratio,curve = openR4ACSV('../build/aaaa_ratio_' + str(i) + '.csv')

    plotR4a(d,ratio,curve,ylim=[0,max(ratio)])
```

```python
def plotSync(x,y,labelx,labely,title):
    plt.figure(figsize=(10,6))
    plt.plot(x,y,'bo-')
    plt.xlabel(labelx)
    plt.ylabel(labely)          
    # plt.legend()
    plt.title(title)
    plt.grid()    
    plt.show()
```

```python
def openCSV(path):
    with open(path, 'r') as f:
        reader = csv.reader(f)
        header = next(reader)  # ヘッダー行を読み飛ばし

        data = []
        for row in reader:
            data.append(row)

    legends = [i for i in header]
    data = [[float(i) for i in row] for row in data]  # 文字列を数値に変換

    # X軸に時間を、Y軸にデータをセット
    eavf = [row[legends.index('Estimated angular velocity frame')] for row in data]
    mavf = [row[legends.index('Measured angular velocity frame')] for row in data]
    a = [row[legends.index('a-skew')] for row in data]
    b = [row[legends.index('b-offset')] for row in data]
    return eavf,mavf,a,b
```

```python
eavf,mavf,a,b = openCSV('../build/latest_new_sync_table.csv')
plotSync(eavf,mavf,'Estimated angular velocity frame []','Measured angular velocity frame []',"New sync method")
```

```python
plt.figure(figsize=(10,6))
plt.plot(eavf,a,'bo-')
plt.plot(eavf,b)
plt.xlabel('Estimated angular velocity frame')
plt.ylabel('Skew')          
# plt.legend()
plt.title('Skew')
plt.grid()    
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(eavf,b)
plt.xlabel('Estimated angular velocity frame')
plt.ylabel('Offset')          
# plt.legend()
plt.title('Offset')
plt.grid()    
plt.show()
```
